```python
# Your mapping dictionary
targets = {
    "diabetes_risk": {0: 20, 1: 80},
    "hypertension_risk": {0: 20, 1: 50, 2: 80},
    "heart_disease_risk": {0: 20, 1: 80},
    "obesity_risk": {0: 20, 1: 50, 2: 80},
    "cholesterol_imbalance": {0: 20, 1: 80}
}

# Apply mappings in-place
for col, mapping in targets.items():
    df[col] = df[col].map(mapping)
```
