```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/tanya-mac/Desktop/Health Risk Predictor from Lifestyle & Medical History/data/processed/Health_risk_prediction.csv")
```

```python
df.head()
```

```python
df = df.rename(columns={
    "diagnosed_diabetes": "diabetes_risk",
    "hypertension_stage": "hypertension_risk",
    "heart_disease_risk": "heart_disease_risk",
    "obesity_stage": "obesity_risk",
    "cholesterol_issue": "cholesterol_imbalance"
})
```

```python
df.head()
```

```python
df.drop(columns=["diabetes_risk_score", "diabetes_stage"], inplace=True)
```

```python
df.shape
```

```python
print(df["diabetes_risk"].value_counts(),"\n\n") 
print(df["hypertension_risk"].value_counts(),"\n\n") 
print(df["heart_disease_risk"].value_counts(),"\n\n")
print(df["obesity_risk"].value_counts(),"\n\n")
print(df["cholesterol_imbalance"].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

le_hyp = LabelEncoder()
le_obs = LabelEncoder()

df['hypertension_risk'] = le_hyp.fit_transform(df['hypertension_risk'])
df['obesity_risk'] = le_obs.fit_transform(df['obesity_risk'])

df.head()
```

```python
print(df["diabetes_risk"].value_counts(),"\n\n") 
print(df["hypertension_risk"].value_counts(),"\n\n") 
print(df["heart_disease_risk"].value_counts(),"\n\n")
print(df["obesity_risk"].value_counts(),"\n\n")
print(df["cholesterol_imbalance"].value_counts())
```

```python
df.to_csv("final_data.csv", index=False)
```
